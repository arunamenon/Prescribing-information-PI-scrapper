### Description: 

FDA provides the exhaustive details of drugs such as their Brand names, Description, Adverse events, Dosage, etc. The details of the different drugs can be accessed from here (https://www.accessdata.fda.gov/scripts/cder/daf/index.cfm). This tool utilises a free api provided by openFDA website(https://open.fda.gov/apis/drug/label/example-api-queries/) to retrieve the information for the drugs passed through it.

## Import Required Packages

```python
import time
import pandas as pd 
from pandas import DataFrame, Series
import requests
import os
import warnings
import datetime
from datetime import date
```

# Enter Filepath to load the list of drugs to be searched and to store Output

```python
drugs_list = pd.read_csv('Input/Drugs_list.csv')
```

```python
output_path = 'Output/'
```

```python
drugs_list = pd.DataFrame(drugs_list)

drugs_list.head()
```

## Using API to get drug information from Accessdata.fda.com. 

Change this to your API generated from the website: https://open.fda.gov/apis/authentication/ (or do not use an API at all)

```python
#Enter API Key below
API_Key = 'X9NvpvVZVTXRJdpfiiDHiS9xIe4ZNy2kM7vPJbDn'
```

```python

start_time = time.time()

drugs_info = pd.DataFrame([]) 

warnings.simplefilter("ignore")

print('Extracting information for the following drugs:')

sleep_time = 2

for i in range(len(drugs_list)):

    
    drug_name = drugs_list.iloc[i,0]
    
    print(drug_name)
    
    
    brand_name_res = requests.get('https://api.fda.gov/drug/label.json?' + 'api_key=' + API_Key + '&search=openfda.brand_name:"' + drug_name + '"&limit=1&sort=effective_time:desc', verify = False)

    generic_name_res = requests.get('https://api.fda.gov/drug/label.json?' + 'api_key=' + API_Key + '&search=openfda.generic_name:"' + drug_name + '"&limit=1&sort=effective_time:desc', verify = False)

    results = pd.DataFrame()
    
    response = ''
    
    if brand_name_res.status_code == 200:
  
        response = brand_name_res
   
    else:
        
        response = generic_name_res
    
    
    if response.status_code == 200:
        
        data = response.json()

        results = pd.DataFrame(data['results'])
    

    results['Row_id'] = i
    
    drugs_info = drugs_info.append(results)

    if i % 10 == 0 and i != 0:
        time.sleep(sleep_time)
       
    
warnings.simplefilter("default")


print('Process time:', round((time.time() - start_time),1), ' seconds')
```

## Final list of drugs with details from Accessdata.fda.com

```python
drugs_list['Row_id'] = drugs_list.index
```

```python
drug_master = pd.merge(drugs_list, drugs_info, how = 'left', on = 'Row_id')
```

```python
drug_master.head()
```

```python
drug_master.drop(columns = ['Row_id'], inplace = True)

writer = pd.ExcelWriter(output_path + str(date.today()) +'_Drug Info.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
drug_master.to_excel(writer, sheet_name='Drug Info')

writer.save()
print('Completed writing the drug information in an excel')
```
